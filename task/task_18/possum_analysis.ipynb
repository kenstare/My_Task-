### This for the own data analysis of the possum data set for mean,median,mode,Range,variance,standard deviation and outlier detection.


```python
import numpy as np
import pandas as pd
import scipy.stats as stats
```

```python
# Load dataset 
df = pd.read_csv("possum.csv")
df.head(20)
```

```python
d = df['hdlngth'].dropna() # drop the missing values if any
print(d) 
```

```python
d.info()
```

```python
d.describe()
```

**Measures of centre**

```python
# Mean
Mean_d = d.mean()
print(Mean_d)
```

```python
# Median
median_d = d.median()
print(median_d)
```

```python
# Mode
mode_d = stats.mode(d, nan_policy="omit")
print(mode_d)
```

**Measures of spread**

```python
# Range
range_d = d.max() - d.min()
print(range_d)
```

```python
### Variance
variance_d = d.var()
print(variance_d)
# or
#var_d = d.var(ddof=1)
#print(var_d)
```

**Standard deviation**

```python
# Standard deviation
std_d = d.std(ddof=1)
print(std_d)
```

**Outlier detection**

```python
q1 = d.quantile(0.25)
q3 = d.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
outliers = d[(d < lower_bound) | (d > upper_bound)]
print(outliers.head())
```

```python
z = (d - Mean_d) / std_d
outliers_z = d[np.abs(z) > 3]
print(z)
print(outliers_z)
```

## This for the each column analysis

```python
import numpy as np
import pandas as pd
```

```python
possum = pd.read_csv("possum.csv")
possum.head()
```

```python
possum.info()
```

```python
possum.describe()

```

```python
possum_data = possum.copy()
```

```python
possum_data.drop("Pop", axis = 1, inplace = True)
possum_data.head(1)
```

**The Center of Measure**

```python
possum_data.head(20)
```

```python

# 1. compute the mode of the "eye" column
possums_case_mode = possum_data["eye"].mode()
print("Mode of 'eye' column:")
print(possums_case_mode)

# 2. Value counts of "eye" column
possums_case_mod = possum_data["eye"].value_counts()
print("\nValue counts of 'eye' column:")
print(possums_case_mod)

# 3. Sorted value counts of "earconch" column
print("\nSorted value counts of 'earconch' column:")
print(possum_data['earconch'].value_counts().sort_values())

# 4. Mode of "earconch" column
possums_case_mode = possum_data['earconch'].mode()
print("\nMode of 'earconch' column:")
print(possums_case_mode)
```

```python
# Define the columns, their labels, and units
columns_info = {
    "case":       ("Case (Observation Number)", ""),
    "site":       ("Site number where the possum was trapped", ""),
    "age":        ("Age", ""),
    "hdlngth":    ("Head length", "mm"),
    "skullw":     ("Skull width", "mm"),
    "totlngth":   ("Total length", "cm"),
    "taill":      ("Tail length", "cm"),
    "footlgth":   ("Foot length", "cm"),
    "earconch":   ("Ear-conch", "mm"),
    "eye":        ("Eye distance from medial to lateral canthus", "mm"),
    "chest":      ("Chest girth", "cm"),
    "belly":      ("Belly girth", "cm")
}

# Loop through each column and compute stats
for col, (description, unit) in columns_info.items():
    print(f"Data on {description} ({unit})" if unit else f"Data on {description}")

    # Use nan-safe functions where appropriate
    col_data = possum_data[col]
    mean_val = np.nanmean(col_data)
    median_val = np.nanmedian(col_data)
    mode_val = col_data.mode()

    # Print results
    print(f"Mean: {mean_val:.1f}{unit}")
    print(f"Median: {median_val:.1f}{unit}")
    print("Mode:")
    print(f"{mode_val.to_list()} {unit}\n")
```

**Measure of Spread**

```python
# Dictionary with columns and their descriptions
columns_info = {
    "case": "Case",
    "site": "Site",
    "age": "Age",
    "hdlngth": "Head Length",
    "skullw": "Skull Width",
    "totlngth": "Total Length",
    "taill": "Tail Length",
    "footlgth": "Foot Length",
    "earconch": "Ear-conch Length",
    "eye": "Eye Distance from Medial Canthus",
    "chest": "Chest",
    "belly": "Belly"
}

# Loop through each column
for col, description in columns_info.items():
    col_data = possum_data[col].dropna()  # Drop NaNs if any

    # Actual range
    col_range = col_data.max() - col_data.min()

    # Approximated values based on range
    approx_std = col_range / 4
    approx_var = approx_std ** 2

    # Print results
    print(f"<=== Data on {description} ===>")
    print(f"Range: {col_range:.1f}")
    print(f"Approx. Variance (range²/16): {approx_var:.1f}")
    print(f"Approx. Std. Deviation (range/4): {approx_std:.1f}\n")
```
