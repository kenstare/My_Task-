```python
work['Year'] = work['Date'].dt.year
work['Month_Name'] = work['Date'].dt.month_name()
work = work.drop(columns=['Month_Name'])
print(work)
```

```python
work['Price'] = work['Price'].astype(str).str.strip().astype(float)
print(work)
```

```python
print(f"duplicates: {work.duplicated().sum()}")
```

```python
work['Revennue'] = work['Price'] * work['Units']
print(work)
```

```python
work.groupby('Date')['Revennue'].sum().plot()
```

```python
work.groupby('Sales Agent')['Revennue'].sum().nlargest(5)
```

```python
pivot = pd.pivot_table(work, values='Revennue', index="Products", columns="Year", aggfunc="sum")
pivot
```

```python
# customer analysis
print("Top 5 Sales Agent")
work.groupby('Sales Agent')['Revennue'].sum().nlargest(5)
```

```python
top_agent = pd.pivot_table(work, values='Revennue', index='Sales Agent', columns='Year', aggfunc='sum')
print("Agent Sales per Year")
top_agent
```

```python
# monthly / year sales
pd.pivot_table(work, values='Revennue', index='Sales Agent', columns='Year', aggfunc='sum')
```

```python
work.groupby('Date')['Revennue'].sum().plot(color='black')
```

```python
# Product Analysis
work.groupby('Products')['Units'].sum().plot(kind='bar', color='red', title="Product Analysis")
```

```python
# Most Proftable Products
work.groupby('Products')['Revennue'].sum().plot(kind='bar', color='green', title="Product Revenue Analysis")

# Branch Analysis
```

```python
# Branch Analysis
branch_summary = work.groupby('Branch')[['Revennue', 'Units', 'Products']].sum()
branch_summary
```
