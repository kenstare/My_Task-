### Implementing the forward function

```python
class Perceptron:
    def __init__(self, num_features):
        self.num_features = num_features
        self.weights = [0. for _ in range(num_features)]
        self.bias = 0

    def forward(self, x):
        weighted_sum_z = self.bias
        for i, _ in enumerate(self.weights):
            weighted_sum_z += x[i] * self.weights[i]

        if weighted_sum_z > 0:
            prediction = 1
        else:
            prediction = 0

        return prediction

ppn = Perceptron(num_features = 2) # Re-initialize with 2 features
x = [1.23, 2.13]
ppn.forward(x)
```

### Training the model

```python
def train(model, X_train, y_train, epochs):
    # Training Loop
    for epoch in range(epochs):
        error_count = 0

        for x, y in zip(X_train, y_train):
            error = model.forward(x)
        print(f"Epoch {epoch + 1} errors {error_count}")
    return error 
            

        
```

```python
ppn = Perceptron(num_features = 2)
train(ppn, features, targets, epochs = 5)
```

### Evaluating the model

```python
def compute_accuracy(model, features, targets):
    correct = 0.0

    for x, y in zip(features, targets):
        prediction = model.forward(x)
        correct += int(prediction == y)

    return correct / len(targets)
```

```python
train_acc = compute_accuracy(ppn, features, targets)
print("Model Accuracy:", train_acc)
```

### Plotting the Decision Boundary

```python
def plot_boundary(model):

    w1, w2 = model.weights[0], model.weights[1]
    b = model.bias

    # z = w1*x1 + w2*x2 + b = 0
    # => x2 = (-w1*x1 - b) / w2

    min_non = 1/10000000
    x1_min = -20
    x2_min = (-(w1 * x1_min) - b) / (w2 + min_non)

    x1_max = 20
    x2_max = (-(w1 * x1_max) - b) / (w2 + min_non)

    return x1_min, x1_max, x2_min, x2_max
```

```python
x1_min, x1_max, x2_min, x2_max = plot_boundary(ppn)

plt.plot(
    features[targets == 0, 0],
    features[targets == 0, 1],
    marker="D",
    markersize=10,
    linestyle="",
    label="Class 0",
)
plt.plot(
    features[targets == 1, 0],
    features[targets == 1, 1],
    marker="^",
    markersize=13,
    linestyle="",
    label="Class 1",
)

# Plot the decision boundary
plt.plot([x1_min, x1_max], [x2_min, x2_max], color="k")

plt.legend(loc=2)
plt.xlim([-2, 2])
plt.ylim([-2, 2])
plt.xlabel("Feature $x_1$", fontsize=12)
plt.ylabel("Feature $x_2$", fontsize=12)
plt.grid()
plt.show()
```
